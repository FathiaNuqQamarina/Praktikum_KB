#### 4. Lakukan Data cleaning pada data dengan nilai null (jika tidak ada nilai null pada dataset, maka buatlah menjadi ada). Ganti nilai null sesuai ketentuan. (bilangan bulat : median/modus, bilangan desimal : mean, tulisan : modus).

```python
#melihat pada dataset apakah terdapat nilai null
print ("Record data dengan nilai null :")
print(df.isna().sum())
```

- pada dataset tidak terdapat nilai null, maka akan dibuat nilai null pada dataset

```python
#membuat nilai null
df.iloc[2,4] = np.nan
df.iloc[3,3] = np.nan
df.head()
```

```python
#data sudah terdapat null
print ("Record data dengan nilai null :")
print(df.isna().sum())

tipeData_pay = df.dtypes['PaymentTier']
print("\nTipe data atribut PaymentTier: ")
print(tipeData_pay)

tipeData_age = df.dtypes['Age']
print("\nTipe data atribut Age")
print(tipeData_age)
```

- Karena tipe data yang memiliki nilai null adalah float/desimal maka penggantian null menggunakan stategy mean

```python
#menangani nilai null
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df[["PaymentTier","Age"]] = imputer.fit_transform(df[["PaymentTier","Age"]])

print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

#### 5. Lakukan Data cleaning pada data dengan nilai duplikat. (Jika tidak ada nilai duplikat pada dataset, maka buatlah menjadi ada)

```python
#melihat dataset apakah terdapat nilai duplikat
print ("Record data dengan nilai duplikat :")
print(df.duplicated().sum())
df[df.duplicated()]
```

```python
#menghapus nilai nilai duplikat pada dataset
df.drop_duplicates(inplace=True)
print("Setelah Data Duplikat dihapus :", df.duplicated().sum())
```
